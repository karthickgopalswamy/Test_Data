```
import pandas as pd
import numpy as np

df = pd.read_csv('dummy.csv')

df.head()
```

```
y1 = [2017, 2018]
temp = df.loc[df['year'].isin(y1)]
temp.head()
```

```
m = np.arange(1,13,1)
table1 = temp.loc[((temp['year'] == 2017) & (temp['month'].isin(m[4:]))) | (temp['year'] == 2018) & (temp['month'].isin(m[:4]))]
table1.head()
```

```
y2 = [2016, 2017]
temp2 = df.loc[df['year'].isin(y2)]

table2 = temp2.loc[((temp2['year'] == 2016) & (temp2['month'].isin(m[4:]))) | (temp2['year'] == 2017) & (temp2['month'].isin(m[:4]))]
```

```
table1['NTT 2018'] = table1['sys_pct50'].mean()/table1['sys_SLTT']
table1['IQR 2018'] = (table1['sys_pct75'] - table1['sys_pct25'])/table1['sys_SLTT']
table1.head()
```

```
table2['NTT 2017'] = table2['sys_pct50'].mean()/table2['sys_SLTT']
table2['IQR 2017'] = (table2['sys_pct75'] - table2['sys_pct25'])/table2['sys_SLTT']
table2.head()
```

```
table1['TOD'] = table1['TOD'].astype('str')
table2['TOD'] = table2['TOD'].astype('str')
```

```
replacements = {
   'TOD': {
      r'5': 'AM', r'6': 'Midday', r'7': 'PM', r'2': 'Weekend'}
}

table1.replace(replacements, regex=True, inplace=True)
table2.replace(replacements, regex=True, inplace=True)

table2.head()
```

```
t1 = table2.pivot_table(index='SYSTEM', columns = 'TOD', values = ['NTT 2017','IQR 2017'])
```

```
t2 = table1.pivot_table(index='SYSTEM', columns = 'TOD', values = ['NTT 2018','IQR 2018'])
```

```
sheet = t1.merge(t2,how='inner',on='SYSTEM')

sheet.head()

```

```
sheet.to_excel('output.xlsx')

files.download('output.xlsx')
```

```
output = sheet.stack(level=0)
output.head()
```

```
output.to_excel('output_updated.xlsx')

files.download('output_updated.xlsx')
```

```
output['Weighted Sum'] = output['AM']*0.189 + output['Midday']*0.210 + output['PM']*0.220 + output['Weekend']*0.190
output.head()
```

```
output.reset_index(inplace=True)
output.head()
```
